```python
results_sy = requests.get(url_sy).json()
results_sy
```

```python
venus_mel = results_mel['response']['groups'][0]['items']
df_mel=json_normalize(venus_mel)
df_mel
```

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =df_mel.loc[:, filtered_columns]

# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)

# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues.head()
```

```python
venus_sy = results_sy['response']['groups'][0]['items']
df_sy=json_normalize(venus_sy)
df_sy
```

```python
# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
sy_venues =df_sy.loc[:, filtered_columns]

# filter the category for each row
sy_venues['venue.categories'] = sy_venues.apply(get_category_type, axis=1)

# clean columns
sy_venues.columns = [col.split(".")[-1] for col in sy_venues.columns]

sy_venues.head()
```

## Methodology

```python
sy_most=sy_venues['categories'].value_counts().to_frame()
mel_most=nearby_venues['categories'].value_counts().to_frame()
```

```python
mel_most=mel_most.iloc[0:10]
```

```python
mel_most=mel_most.reset_index()
mel_most
```

```python
sy_most=sy_most.iloc[0:10]
sy_most=sy_most.reset_index()
sy_most
```

```python
most=pd.merge(mel_most,sy_most, how='outer', on='index')
most.columns=['categories','mel','sy']
most.fillna(0,inplace=True)
most
```

```python
most.plot(kind='bar', x='categories', figsize=(20,8),width=0.8)
plt.title('most venues in melbourne and sydney')
plt.ylabel('number')
```

## visualization

```python
# one hot encoding
mel_onehot = pd.get_dummies(nearby_venues[['categories']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
mel_onehot['name'] = nearby_venues['name'] 

# move neighborhood column to the first column
fixed_columns = [mel_onehot.columns[-1]] + list(mel_onehot.columns[:-1])
mel_onehot = mel_onehot[fixed_columns]

mel_onehot.head()
```

```python
mel_map = folium.Map(location=[latitude_mel, longitude_mel], zoom_start=13)
for lat, lng, label in zip(nearby_venues.lat, nearby_venues.lng, nearby_venues.categories):
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        color='blue',
        popup=label,
        fill = True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(mel_map)

# display map
mel_map
```

```python
sy_map = folium.Map(location=[latitude_sy, longitude_sy], zoom_start=13)
for lat, lng, label in zip(sy_venues.lat, sy_venues.lng, sy_venues.categories):
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        color='blue',
        popup=label,
        fill = True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(sy_map)

# display map
sy_map
```
